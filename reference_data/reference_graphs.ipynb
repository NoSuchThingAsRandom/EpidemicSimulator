```python
import pandas
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy
import datetime as dt
plt.rcParams["figure.figsize"] = (25, 25)
```

```python
york_cases=pandas.read_csv("York/cases.csv")
york_cases.columns = york_cases.columns.str.replace(' ', '')
york_cases.columns
```

```python
dates=[dt.datetime.strptime(d,'%Y-%m-%d').date() for d in york_cases["date"]]


plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=28))
plt.plot(dates,york_cases["newCasesBySpecimenDate"])
plt.gcf().autofmt_xdate()

#plt.plot(york_cases["newCasesBySpecimenDate"])
#plt.xticks(range(0,len(york_cases["newCasesBySpecimenDate"])),york_cases["date"])
```
